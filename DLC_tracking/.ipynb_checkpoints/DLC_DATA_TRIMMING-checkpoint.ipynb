```python
import h5py
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import matplotlib.patches as patches
path = r'C:\Users\maksymilianm\Dropbox (UCL - SWC)\Project_spiders\Analysis\confined_shade_arena_exploration\06_02_20_sp13DLC_resnet50_large_spidersJan27shuffle1_1030000.h5'

data = pd.read_hdf(path)
#for key in data:
 #   print(key)
centre_x, centre_y, centre_likelyhood = (data['DLC_resnet50_large_spidersJan27shuffle1_1030000']['centre']['x'], 
                                        data['DLC_resnet50_large_spidersJan27shuffle1_1030000']['centre']['y'],
                                        data['DLC_resnet50_large_spidersJan27shuffle1_1030000']['centre']['likelihood'])
head_x, head_y = (data['DLC_resnet50_large_spidersJan27shuffle1_1030000']['head']['x'], 
                   data['DLC_resnet50_large_spidersJan27shuffle1_1030000']['head']['y'])
tail_x, tail_y = (data['DLC_resnet50_large_spidersJan27shuffle1_1030000']['tail']['x'], 
                   data['DLC_resnet50_large_spidersJan27shuffle1_1030000']['tail']['y'])
#AR leg_x, AR leg_y = (data['DLC_resnet50_large_spidersJan27shuffle1_1030000']['AR leg']['x'], 
#             data['DLC_resnet50_large_spidersJan27shuffle1_1030000']['AR leg']['y'])
#AL leg_x, AL leg_y = (data['DLC_resnet50_large_spidersJan27shuffle1_1030000']['AL leg']['x'], 
#            data['DLC_resnet50_large_spidersJan27shuffle1_1030000']['AL leg']['y'])
#PR leg_x, PR leg_y = (data['DLC_resnet50_large_spidersJan27shuffle1_1030000']['PR leg_x']['x'], 
#             data['DLC_resnet50_large_spidersJan27shuffle1_1030000']['PR leg_x']['y'])
#PL leg_x, PL leg_y = (data['DLC_resnet50_large_spidersJan27shuffle1_1030000']['PL leg']['x'], 
#             data['DLC_resnet50_large_spidersJan27shuffle1_1030000']['PL leg']['y'])
def compute_distance(xs,ys):
    
    dx = np.diff(xs, axis=0)
    dy = np.diff(ys, axis=0)
    
    distances = np.hypot(dx,dy)
    
    return distances
```

```python
distance_per_frame = compute_distance(centre_x, centre_y)
fig, ax = plt.subplots(figsize=(15,5))
ax.set_ylim(0, 500)
#Adjust the x-axis label range and interval to better see which parts need to be excluded
plt.xticks(np.arange(0,520000, step=6000), rotation = 'vertical')
plt.plot(distance_per_frame)
```
